```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully")
```

```python
df = pd.read_excel(r"C:\Users\Hhari\OneDrive\Documents\Project_2_Retail_Profit_Intelligence\data\Sample - Superstore.csv.xlsx")

print("Data loaded from enterprise project folder successfully")

df.head()
```

```python
# ================================
# ENTERPRISE FEATURE ENGINEERING
# ================================

# Profit Margin %
df['Profit Margin %'] = (df['Profit'] / df['Sales']) * 100

# Time Intelligence Columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day

# Business Intelligence Columns
df['Sales Category'] = df['Sales'].apply(
    lambda x: 'High Sales' if x > 500 else 'Medium Sales' if x > 100 else 'Low Sales'
)

df['Profit Category'] = df['Profit'].apply(
    lambda x: 'Loss' if x < 0 else 'Low Profit' if x < 50 else 'High Profit'
)

print("Enterprise feature engineering completed successfully")

df.head()
```

```python
# ==========================================
# EXECUTIVE PROFIT INTELLIGENCE ANALYSIS
# ==========================================

# Total Sales
total_sales = df['Sales'].sum()

# Total Profit
total_profit = df['Profit'].sum()

# Average Profit Margin
avg_profit_margin = df['Profit Margin %'].mean()

# Best Performing Region
best_region = df.groupby('Region')['Profit'].sum().idxmax()

# Worst Performing Region
worst_region = df.groupby('Region')['Profit'].sum().idxmin()

# Best Category
best_category = df.groupby('Category')['Profit'].sum().idxmax()

# Worst Category
worst_category = df.groupby('Category')['Profit'].sum().idxmin()

# Print Executive Summary
print("========== EXECUTIVE BUSINESS SUMMARY ==========")

print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}%")

print(f"Best Performing Region: {best_region}")
print(f"Worst Performing Region: {worst_region}")

print(f"Best Performing Category: {best_category}")
print(f"Worst Performing Category: {worst_category}")

print("===============================================")
```

```python
# ==========================================
# EXECUTIVE VISUALIZATION DASHBOARD
# ==========================================

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Profit by Region
plt.figure(figsize=(8,5))
region_profit = df.groupby('Region')['Profit'].sum().sort_values()
sns.barplot(x=region_profit.index, y=region_profit.values)
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()


# Profit by Category
plt.figure(figsize=(8,5))
category_profit = df.groupby('Category')['Profit'].sum().sort_values()
sns.barplot(x=category_profit.index, y=category_profit.values)
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()


# Sales Trend Over Time
plt.figure(figsize=(10,5))
sales_trend = df.groupby('Year')['Sales'].sum()
sns.lineplot(x=sales_trend.index, y=sales_trend.values, marker="o")
plt.title("Sales Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```
